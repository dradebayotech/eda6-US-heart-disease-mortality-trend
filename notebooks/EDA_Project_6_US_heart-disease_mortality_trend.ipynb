## **Normalizing and Renaming column names to snake_case**

```python
# Normalize column names to snake_case
heart_disease_Mortality.columns = heart_disease_Mortality.columns.str.lower().str.replace(' ', '_')

# Rename the columns to a nicer name.
heart_disease_Mortality = heart_disease_Mortality.rename(columns={
    'year':'year',
    'locationabbr':'location_abbr',
    'locationdesc':'location_desc',
    'geographiclevel':'geographic_level',
    'data_value':'data_value',
    'data_value_unit':'data_value_unit',
    'data_value_type':'data_value_type',
    'data_value_footnote':'data_value_footnote',
    'stratificationcategory1' : 'stratification_category1',
    'stratification1':'stratification1',
    'stratificationcategory2':'stratification_category2',
    'stratification2':'stratification2',
    'topicid':'topic_id',
    'locationid':'location_id',
    'y_lat':'y_lat',
    'x_lon':'x_lon',
    'georeference':'georeference'
})


# Display the new column names to confirm
print("New column names after normalization:")
print(heart_disease_Mortality.columns)
```

## **Coerce (Convert) Numeric Column and Create Flags**

```python
# Coerce 'data_value' to numeric, converting errors to NaN
heart_disease_Mortality['data_value'] = pd.to_numeric(heart_disease_Mortality['data_value'], errors='coerce')

# Create flags
heart_disease_Mortality['is_smoothed'] = heart_disease_Mortality['data_value_type'].str.contains('Spatially Smoothed', na=False)
heart_disease_Mortality['is_county'] = heart_disease_Mortality['geographic_level']=='County'
heart_disease_Mortality['is_state']  = heart_disease_Mortality['geographic_level']=='State'
heart_disease_Mortality['is_nation'] = heart_disease_Mortality['geographic_level']=='Nation'
heart_disease_Mortality['insufficient_data_flag'] = heart_disease_Mortality['data_value_footnote'].fillna('').str.contains('Insufficient Data')

# Check the first five record to check the flags
heart_disease_Mortality.head()
```

## **Handling the Insufficient Data rows**

```python
# Check the number of rows with 'Insufficient Data' flag
print(heart_disease_Mortality['insufficient_data_flag'].sum())

# Display the first few rows of the 'Insufficient Data' flag
print(heart_disease_Mortality[['data_value_footnote', 'insufficient_data_flag']].head())

# Create a new DAtaFrame to Filter the rows flagged 'Insufficient Data'
heart_disease_Mortality_clean = heart_disease_Mortality[~heart_disease_Mortality['insufficient_data_flag']].copy()

# Display the first few rows of filtered/clean DataFrame
print(heart_disease_Mortality_clean[['data_value_footnote', 'insufficient_data_flag']].head())
```

## **Standardize location_id (FIPS) into zero-padded strings**

```python
# Standardize location_id to a zero-padded string
# Assuming Federal Information Processing Standards (FIPS) codes are typically 5 digits for counties and 2 for states/territories.
# We'll pad to 5 digits to accommodate county codes.
heart_disease_Mortality_clean['location_id_str'] = heart_disease_Mortality_clean['location_id'].astype(str).str.zfill(5)

# Display the first few rows with the new column to confirm
print(heart_disease_Mortality_clean[['location_id', 'location_id_str']].head())
```

## **Remove Duplicates and Check Inconsistencies**

```python
# Check for duplicate rows
print(f"Number of duplicate rows before removal: {heart_disease_Mortality_clean.duplicated().sum()}")

# Remove duplicate rows
heart_disease_Mortality_clean.drop_duplicates(inplace=True)

# Check for duplicate rows after removal
print(f"Number of duplicate rows after removal: {heart_disease_Mortality_clean.duplicated().sum()}")
```

# **EDA Question 2: Heart Disease Mortality by Sex**

## **QUESTION 2: How do men and women compare in heart disease mortality across geographies?**

*   **Analysis:** Paired comparisons at the same location (male vs female `data_value`), compute male/female rate ratio and absolute difference.
*   **Plots/tests:** Paired scatter (y=x line), violin/boxplot by sex, map of male:female rate ratio. Use Wilcoxon signed-rank (if non-normal).
*   **Medical relevance:** Quantify sex differences and geographic heterogeneity.

```python
# Filter data for Male and Female stratifications at the county level (using smoothed data)
male_mortality = heart_disease_Mortality_clean[
    (heart_disease_Mortality_clean['geographic_level'] == 'County') &
    (heart_disease_Mortality_clean['is_smoothed'] == True) &
    (heart_disease_Mortality_clean['stratification1'] == 'Male') &
    (heart_disease_Mortality_clean['stratification2'] == 'Overall') # Assuming 'Overall' race/ethnicity for comparison
].copy()

female_mortality = heart_disease_Mortality_clean[
    (heart_disease_Mortality_clean['geographic_level'] == 'County') &
    (heart_disease_Mortality_clean['is_smoothed'] == True) &
    (heart_disease_Mortality_clean['stratification1'] == 'Female') &
    (heart_disease_Mortality_clean['stratification2'] == 'Overall') # Assuming 'Overall' race/ethnicity for comparison
].copy()

# Merge male and female data on location_id_str to create paired data
paired_mortality = pd.merge(male_mortality, female_mortality,
                            on=['location_id_str', 'location_desc', 'location_abbr', 'geographic_level', 'year'],
                            suffixes=('_male', '_female'))

# Compute rate ratio and absolute difference
paired_mortality['rate_ratio_male_female'] = paired_mortality['data_value_male'] / paired_mortality['data_value_female']
paired_mortality['rate_difference_male_female'] = paired_mortality['data_value_male'] - paired_mortality['data_value_female']

# Display the first few rows of the paired data with the new calculations
print("Paired Mortality Data (Male vs Female) with Rate Ratio and Difference:")
display(paired_mortality[['location_desc', 'data_value_male', 'data_value_female', 'rate_ratio_male_female', 'rate_difference_male_female']].head())
```

### **Let's visualize the comparison between male and female heart disease mortality rates across different counties**

```python
# Create a paired scatter plot of male vs female mortality rates
plt.figure(figsize=(8, 8))
sns.scatterplot(x='data_value_female', y='data_value_male', data=paired_mortality, alpha=0.6)

# Add a y=x line for reference
plt.plot([0, paired_mortality[['data_value_female', 'data_value_male']].max().max()],
         [0, paired_mortality[['data_value_female', 'data_value_male']].max().max()],
         color='red', linestyle='--')

plt.xlabel('Female Age-Adjusted Mortality Rate (per 100,000)')
plt.ylabel('Male Age-Adjusted Mortality Rate (per 100,000)')
plt.title('Paired Scatter Plot of Male vs Female Heart Disease Mortality Rates by County (2019-2021 Avg)')
plt.grid(True)
plt.show()
```

```python
# Calculate average mortality rate per state in the paired data
state_avg_mortality = paired_mortality.groupby('location_abbr')[['data_value_male', 'data_value_female']].mean()

# Calculate the average of male and female mortality to rank states
state_avg_mortality['avg_mortality'] = (state_avg_mortality['data_value_male'] + state_avg_mortality['data_value_female']) / 2

# Get the top 20 states by average mortality
top_20_states = state_avg_mortality.sort_values(by='avg_mortality', ascending=False).head(20).index.tolist()

# Filter the paired_mortality DataFrame to include only counties in the top 20 states
paired_mortality_top_states = paired_mortality[paired_mortality['location_abbr'].isin(top_20_states)].copy()

# Create a scatter plot for the top 20 states, colored by state
plt.figure(figsize=(14, 12))
sns.scatterplot(x='data_value_female', y='data_value_male', data=paired_mortality_top_states, alpha=0.6, hue='location_abbr', palette='tab20') # Use a palette with more distinct colors

# Add a y=x line for reference
plt.plot([0, paired_mortality_top_states[['data_value_female', 'data_value_male']].max().max()],
         [0, paired_mortality_top_states[['data_value_female', 'data_value_male']].max().max()],
         color='red', linestyle='--')

plt.xlabel('Female Age-Adjusted Mortality Rate (per 100,000)')
plt.ylabel('Male Age-Adjusted Mortality Rate (per 100,000)')
plt.title('Paired Scatter Plot of Male vs Female Heart Disease Mortality Rates by County in Top 20 States (2019-2021 Avg)')
plt.grid(True)
plt.show()
```

Based on the analysis of age-adjusted heart disease mortality rates for males and females across counties (2019-2021 3-yr avg, smoothed data):

# **EDA Question 3: Heart Disease Mortality by Race/Ethnicity**

## **QUESTION 3: How do mortality rates vary across race/ethnicity groups — and where are disparities largest?**

*   **Analysis:** Compute group-wise summaries (median, IQR) by `Stratification2`, and compute rate ratios vs White group.
*   **Plots/tests:** Boxplots by race; maps for specific race groups (e.g., Black vs White); Kruskal-Wallis + post-hoc Dunn tests.
*   **Caveat:** Many race × county cells are suppressed → interpret sparse regions carefully.

```python
# Filter data for county-level, smoothed data and different race/ethnicity stratifications
# Excluding 'Overall' for this analysis to focus on specific race/ethnicity groups
race_ethnicity_mortality = heart_disease_Mortality_clean[
    (heart_disease_Mortality_clean['geographic_level'] == 'County') &
    (heart_disease_Mortality_clean['is_smoothed'] == True) &
    (heart_disease_Mortality_clean['stratification1'] == 'Overall') & # Using Overall sex for race/ethnicity comparison
    (heart_disease_Mortality_clean['stratification2'] != 'Overall') # Exclude Overall race/ethnicity
].copy()

# Compute group-wise summaries (median, IQR) by Stratification2 (Race/Ethnicity)
race_ethnicity_summary = race_ethnicity_mortality.groupby('stratification2')['data_value'].agg(['median', lambda x: x.quantile(0.75) - x.quantile(0.25)]).rename(columns={'<lambda_0>': 'IQR'})

print("Summary Statistics (Median and IQR) of Heart Disease Mortality Rates by Race/Ethnicity:")
display(race_ethnicity_summary)
```

```python
# Create boxplots of mortality rates by race/ethnicity with updated styling
plt.figure(figsize=(12, 8))
sns.boxplot(x='data_value', y='stratification2', data=race_ethnicity_mortality, orient='h', palette='Set1')
plt.xlabel('Age-Adjusted Heart Disease Mortality Rate (per 100,000)', fontsize=12, fontweight='bold')
plt.ylabel('Race/Ethnicity', fontsize=14, fontweight='bold')
plt.title('Distribution of Heart Disease Mortality Rates by Race/Ethnicity (2019-2021 Avg, Smoothed County Data)', fontsize=16, fontweight='bold')
plt.yticks(weight='bold')
plt.xticks(weight='bold')
plt.tight_layout()
plt.show()
```

## **Answer to Question 3: Heart Disease Mortality by Race/Ethnicity**

Based on the analysis of age-adjusted heart disease mortality rates across different race and ethnicity groups (2019-2021 3-yr avg, smoothed county data):

*   There are notable variations in age-adjusted heart disease mortality rates across different race and ethnicity groups.
*   Based on the median mortality rates, the **Black** and **Native Hawaiian or Other Pacific Islander** groups appear to have the highest mortality rates, followed by **White** and **American Indian or Alaska Native** groups.
*   The **Asian** and **More than one race** groups show the lowest median mortality rates.
*   The boxplot visually supports these observations, showing the distribution and spread of mortality rates for each group. The boxes and whiskers indicate the range and central tendency of the rates within each category.
*   The IQR values provide a measure of the spread of mortality rates within each group. Some groups show a wider spread (larger IQR) than others, indicating greater variability in mortality rates within those populations across counties.

# **EDA Question 4: Heart Disease Mortality Hotspots by Race and Sex**

## **QUESTION 4: Where do specific race × sex subgroups show especially high rates (hotspots)?**

*   **Analysis:** Filter to (race,sex) combinations and map rates — e.g., Black males, AI/AN overall.
*   **Plots:** Small-multiples county maps by subgroup; LISA/local cluster maps to find localized hotspots.

```python
# Filter data for the specified subgroups at the county level (using smoothed data)
subgroups_to_map = {
    'AI/AN Overall': {'stratification1': 'Overall', 'stratification2': 'American Indian or Alaska Native'},
    'White Males': {'stratification1': 'Male', 'stratification2': 'White'},
    'Black Females': {'stratification1': 'Female', 'stratification2': 'Black'},
    'White Females': {'stratification1': 'Female', 'stratification2': 'White'},
    'Black Males': {'stratification1': 'Male', 'stratification2': 'Black'}, # Added Black Males
    'Nat Hawaiian/Other PI Males': {'stratification1': 'Male', 'stratification2': 'Native Hawaiian or Other Pacific Islander'}, # Added Nat Hawaiian/Other PI Males
    'Nat Hawaiian/Other PI Females': {'stratification1': 'Female', 'stratification2': 'Native Hawaiian or Other Pacific Islander'} # Added Nat Hawaiian/Other PI Females
}

# Reuse the merged_gdf from Question 1, but filter for the specific subgroups
# We need to re-merge for each subgroup as the data_value column will change
# Alternatively, we can filter the original heart_disease_Mortality_clean and then merge

# Let's filter the original cleaned data first for simplicity
filtered_subgroups_mortality = heart_disease_Mortality_clean[
    (heart_disease_Mortality_clean['geographic_level'] == 'County') &
    (heart_disease_Mortality_clean['is_smoothed'] == True) &
    (heart_disease_Mortality_clean['stratification1'].isin([s['stratification1'] for s in subgroups_to_map.values()])) &
    (heart_disease_Mortality_clean['stratification2'].isin([s['stratification2'] for s in subgroups_to_map.values()]))
].copy()

# Merge with the counties geodataframe (assuming counties_gdf is already loaded from Question 1)
# Ensure the FIPS codes in both dataframes have the same format (5-digit string)
# counties_gdf['GEOID'] = counties_gdf['GEOID'].astype(str).str.zfill(5) # Already done in Q1
# filtered_subgroups_mortality['location_id_str'] = filtered_subgroups_mortality['location_id'].astype(str).str.zfill(5) # Already done in cleaning

merged_subgroups_gdf = counties_gdf.merge(filtered_subgroups_mortality, left_on='GEOID', right_on='location_id_str', how='left')

# Calculate the number of rows and columns for the subplot grid
num_subgroups = len(subgroups_to_map)
n_cols = 3 # Number of columns in the grid
n_rows = (num_subgroups + n_cols - 1) // n_cols # Calculate the number of rows needed

# Create small multiples maps
fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(n_cols * 8, n_rows * 6)) # Adjusted figure size
axes = axes.flatten() # Flatten the array of axes for easy iteration

# Remove any unused axes if the number of subgroups is less than n_rows * n_cols
for j in range(num_subgroups, n_rows * n_cols):
    fig.delaxes(axes[j])


for i, (subgroup_name, filters) in enumerate(subgroups_to_map.items()):
    ax = axes[i]

    # Filter the merged geodataframe for the current subgroup
    current_subgroup_gdf = merged_subgroups_gdf[
        (merged_subgroups_gdf['stratification1'] == filters['stratification1']) &
        (merged_subgroups_gdf['stratification2'] == filters['stratification2'])
    ].copy()

    # Plot the choropleth map for the current subgroup
    current_subgroup_gdf.plot(column='data_value', ax=ax, legend=True,
                              cmap='viridis',  # Use a consistent colormap
                              legend_kwds={'label': "Mortality Rate (per 100,000)",
                                           'orientation': "horizontal"})

    ax.set_title(f'Mortality Rate for {subgroup_name}')
    ax.set_axis_off()

plt.tight_layout()
plt.show()
```

## **Answer to Question 4: Heart Disease Mortality Hotspots by Race and Sex**

Based on the small multiples maps visualizing age-adjusted heart disease mortality rates (2019-2021 3-yr avg, smoothed county data) for specific race and sex subgroups:

*   **Black Males and Black Females:** The maps for both Black males and Black females show particularly high mortality rates concentrated in the southeastern United States, aligning with the overall pattern observed in Question 1. There appear to be significant hotspots in states like Mississippi, Louisiana, Alabama, Georgia, and parts of the Carolinas and Virginia. *   **White Males:** The map for White males also indicates elevated mortality rates in the southeastern US, as well as in parts of the Appalachian region and some areas in the Midwest. *   **White Females:** White females tend to have lower mortality rates compared to White males and both Black males and females. The map shows some areas of higher rates in the Southeast and Appalachia, but the intensity and widespread nature of hotspots are less pronounced than in the other groups.

*   **American Indian or Alaska Native (Overall):** The map for the American Indian or Alaska Native population (overall sex) suggests elevated rates in areas with significant AI/AN populations, including parts of the Intermountain West, the Dakotas, and Oklahoma. Data availability might be a factor in the appearance of this map due to suppressed data in many counties.

*   **Native Hawaiian or Other Pacific Islander Males and Females:** The maps for Native Hawaiian or Other Pacific Islander males and females are sparser, likely due to the smaller population size and data suppression in many counties. Where data is available, there are indications of elevated rates in some areas, but it is difficult to identify widespread, consistent hotspots from this visualization alone.

In summary, the analysis by race and sex subgroups reveals that the burden of high heart disease mortality is not evenly distributed.
